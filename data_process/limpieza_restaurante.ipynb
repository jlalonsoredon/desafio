```python
# concatenamos buytype de df_tiendas en Tipo comida sin reducir el dataset de restaurantes
n_rows_before = len(df_restaurantes)

buytype_lookup = (
    df_tiendas[["documentname", "buytype"]]
    .dropna(subset=["documentname"])
)
buytype_lookup["documentname"] = buytype_lookup["documentname"].astype(str).str.strip()
buytype_lookup = (
    buytype_lookup.groupby("documentname", as_index=False)["buytype"]
    .agg(lambda s: " | ".join(dict.fromkeys(str(v).strip() for v in s if pd.notna(v) and str(v).strip())))
)

def concat_unique_types(tipo_actual, tipo_nuevo):
    values = []
    for value in [tipo_actual, tipo_nuevo]:
        if pd.isna(value):
            continue
        for part in str(value).split("|"):
            clean = part.strip()
            if clean and clean not in values:
                values.append(clean)
    return " | ".join(values) if values else pd.NA

def concat_unique_places(place_actual, place_new):
    values = []
    seen = set()
    for value in [place_actual, place_new]:
        if pd.isna(value):
            continue
        for part in str(value).replace("|", ",").split(","):
            clean = part.strip()
            key = clean.lower()
            if clean and key not in seen:
                values.append(clean)
                seen.add(key)
    return ", ".join(values) if values else pd.NA

# left merge: conserva todas las filas de df_restaurantes
df_restaurantes = df_restaurantes.merge(
    buytype_lookup.rename(columns={"documentname": "Nombre", "buytype": "buytype_match"}),
    on="Nombre",
    how="left"
 )

# solo concatenamos en filas comunes (donde hay match)
mask_match = df_restaurantes["buytype_match"].notna()
df_restaurantes.loc[mask_match, "Tipo comida"] = df_restaurantes.loc[mask_match].apply(
    lambda row: concat_unique_types(row.get("Tipo comida"), row.get("buytype_match")),
    axis=1
)

has_tienda = df_restaurantes["Tipo de lugar"].fillna("").str.lower().str.contains(r"\btienda(?:s)?\b", regex=True)
df_restaurantes.loc[mask_match & ~has_tienda, "Tipo de lugar"] = df_restaurantes.loc[
    mask_match & ~has_tienda, "Tipo de lugar"
].apply(lambda x: concat_unique_places(x, "tienda"))

df_restaurantes.drop(columns=["buytype_match"], inplace=True)

n_rows_after = len(df_restaurantes)
if n_rows_after != n_rows_before:
    raise ValueError(f"El dataset se redujo de {n_rows_before} a {n_rows_after} filas")
```

```python
import re
import unicodedata
import requests

n_rows_before = len(df_restaurantes)
url = "https://opendata.euskadi.eus/contenidos/ds_recursos_turisticos/restaurantes_sidrerias_bodegas/opendata/restaurantes.geojson"

resp = requests.get(url, timeout=30)
resp.raise_for_status()
geojson_data = resp.json()

properties = pd.json_normalize([f.get("properties", {}) for f in geojson_data.get("features", [])])

candidate_name_cols = ["documentname", "nameeu", "namees", "name", "Nombre"]
name_col = next((c for c in candidate_name_cols if c in properties.columns), None)
if name_col is None or "restorationtype" not in properties.columns:
    raise KeyError("No se encontraron columnas de nombre/restorationtype en el GeoJSON")

def normalize_name(text):
    text = "" if pd.isna(text) else str(text)
    text = unicodedata.normalize("NFKD", text).encode("ascii", "ignore").decode("utf-8")
    text = re.sub(r"\s+", " ", text).strip().lower()
    return text

geo_types = (
    properties[[name_col, "restorationtype"]]
    .dropna(subset=[name_col])
    .rename(columns={name_col: "nombre_geo"})
)
geo_types["nombre_key"] = geo_types["nombre_geo"].apply(normalize_name)
geo_types = (
    geo_types.groupby("nombre_key", as_index=False)["restorationtype"]
    .agg(lambda s: " | ".join(dict.fromkeys(str(v).strip() for v in s if pd.notna(v) and str(v).strip())))
)

df_restaurantes["nombre_key"] = df_restaurantes["Nombre"].apply(normalize_name)
restoration_map = geo_types.set_index("nombre_key")["restorationtype"]
df_restaurantes["restorationtype_match"] = df_restaurantes["nombre_key"].map(restoration_map)

mask_match = df_restaurantes["restorationtype_match"].notna()
df_restaurantes.loc[mask_match, "Tipo comida"] = df_restaurantes.loc[mask_match].apply(
    lambda row: concat_unique_types(row.get("Tipo comida"), row.get("restorationtype_match")),
    axis=1
)

df_restaurantes.drop(columns=["nombre_key", "restorationtype_match"], inplace=True)

n_rows_after = len(df_restaurantes)
if n_rows_after != n_rows_before:
    raise ValueError(f"El dataset se redujo de {n_rows_before} a {n_rows_after} filas")
```

```python
load_dotenv()
API_KEY = os.getenv("API_KEY_MAPS")
```

```python
def safe_json(response, context="request"):
    """Devuelve JSON seguro; imprime errores de estado/parsing y retorna None si falla."""
    try:
        response.raise_for_status()
    except requests.HTTPError as exc:
        print(f"[HTTP {response.status_code}] {context}: {exc}")
        return None

    try:
        return response.json()
    except ValueError:
        print(f"[JSON ERROR] {context}: respuesta no valida")
        return None


def search_place_id(nombre, municipio, lat, log, web, api_key):
    """Busca Place ID en Google Places. Primero por nombre+municipio y luego por web."""
    payload = {"textQuery": f"{nombre}, {municipio}"}
    if pd.notna(lat) and pd.notna(log):
        payload["locationBias"] = {
            "circle": {
                "center": {"latitude": float(lat), "longitude": float(log)},
                "radius": 500.0
            }
        }

    r = requests.post(
        "https://places.googleapis.com/v1/places:searchText",
        json=payload,
        headers={
            "Content-Type": "application/json",
            "X-Goog-Api-Key": api_key,
            "X-Goog-FieldMask": "places.id"
        },
        timeout=20
    )
    places = (safe_json(r, f"search '{nombre}'") or {}).get("places", [])

    if not places and web:
        print(f"[FALLBACK web] '{nombre}' -> {web}")
        r2 = requests.post(
            "https://places.googleapis.com/v1/places:searchText",
            json={"textQuery": str(web)},
            headers={
                "Content-Type": "application/json",
                "X-Goog-Api-Key": api_key,
                "X-Goog-FieldMask": "places.id"
            },
            timeout=20
        )
        places = (safe_json(r2, f"search web '{nombre}'") or {}).get("places", [])

    return places[0]["id"] if places else None


def get_place_details(place_id, api_key):
    """Obtiene solo la direccion formateada de un Place ID."""
    r = requests.get(
        f"https://places.googleapis.com/v1/places/{place_id}",
        headers={
            "X-Goog-Api-Key": api_key,
            "X-Goog-FieldMask": "formattedAddress"
        },
        timeout=20
    )
    return safe_json(r, f"details '{place_id}'")


def enrich_dataframe(df, api_key, limit=None):
    """Enriquece SOLO la columna Direccion usando Google Places API."""
    if not api_key:
        raise ValueError("API_KEY_MAPS no esta configurada")

    df = df.copy()
    df["Dirección"] = pd.Series(df.get("Dirección", ""), index=df.index, dtype="object")

    subset = df.head(limit) if limit else df
    cont_ok, cont_ko = 0, 0

    for idx, row in subset.iterrows():
        nombre = row.get("Nombre")
        municipio = row.get("Municipio")
        lat = row.get("lat")
        log = row.get("log")
        web = row.get("WEB") if pd.notna(row.get("WEB")) and str(row.get("WEB")).strip() else None

        place_id = search_place_id(nombre, municipio, lat, log, web, api_key)
        if not place_id:
            print(f"[SKIP] sin resultados para '{nombre}'")
            continue

        details = get_place_details(place_id, api_key)
        if not details:
            print(f"[SKIP] sin direccion para '{nombre}'")
            cont_ko += 1
            continue

        df.at[idx, "Dirección"] = details.get("formattedAddress", "")
        cont_ok += 1
        print(f"[OK] '{nombre}' -> {df.at[idx, 'Dirección']} (ok: {cont_ok})")

    print("\n--- RESUMEN ---")
    print(f"Direcciones enriquecidas: {cont_ok} / {len(subset)} | Fallos: {cont_ko}")
    return df
```
